```python
from fefetlab.instruments.visa_session import VisaSession, VisaConfig
from fefetlab.b1500 import B1500, B1500Error

cfg = VisaConfig(
    resource="GPIB0::17::INSTR",
    timeout_ms=10000,
    write_termination="\r\n",
    read_termination="\r\n",
    send_end=True,
)

CH = 1
V_RANGE = 0
V_FORCE = 0.1
I_COMP = 1e-3

with VisaSession(cfg) as s:
    b = B1500(s)

    try:
        print("OPC:", b.opc())
        print("ERRX before:", b.errx())

        b.fmt(5)
        b.cn([CH])

        print(f"Force DV on CH{CH}: vrange={V_RANGE}, V={V_FORCE}, Icomp={I_COMP}")
        b.dv(CH, V_RANGE, V_FORCE, I_COMP)

        i_meas = b.ti(CH)
        print(f"Measured current on CH{CH}: {i_meas}")

        print("ERRX after measure:", b.errx())

    except B1500Error as e:
        print("B1500Error:", e)

    finally:
        try:
            b.dz([CH])
        except Exception as e:
            print("DZ failed:", e)

        try:
            b.cl([CH])
        except Exception as e:
            print("CL failed:", e)
```
